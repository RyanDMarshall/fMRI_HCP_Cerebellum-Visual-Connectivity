```python
import numpy as np
import nibabel as nb
import csv
import pdb
```

```python
def read_by_tokens(fn):
    fileobj=open(fn)
    for line in fileobj:
        for token in line.split():
            yield token
            
def save_dict_to_csv(data, filename):
    # Extract fieldnames from first dictionary in list
    fieldnames = data[0].keys()
    # Open file for writing in binary mode
    with open(filename, 'w', newline='') as csvfile:
        # Create CSV writer object
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        # Write header row
        writer.writeheader()
        # Write data rows
        for row in data:
            writer.writerow(row)
            
cb_retmaps = nb.load('/projectnb/somerslab/ryanmars/atlases/cerebellum/vanEsRetinotopy/cerebellum_retmaps.nii')
base_dir=f'/projectnb/somerslab/ryanmars/projects/CrossTaskAnalysis/proj_allrest/results/group/HCP_SUIT'
roi_fn=f'{base_dir}/rois-sorted.txt'
roi_list=read_by_tokens(roi_fn)
nsub=166
```

```python
output = list()
output_fn=f'{base_dir}/retinotopic_tstats.csv'

for roi in roi_list:
    for corr in ("pearson","partial"):
        for hemi in ("lh","rh","bilat"):
            if corr=="partial" and hemi=="bilat":
                continue
            sub_dir=f'{base_dir}/group_tstat/{corr}'
            fn_base=f'perm_vsm4_{roi}_{hemi}_{corr}_tstat1.nii'
            fn=f'{sub_dir}/{fn_base}'
            vol=nb.load(fn)
            for val in np.unique(cb_retmaps.get_fdata()):
                vol_fd=vol.get_fdata()
                mask=(cb_retmaps.get_fdata()==val)
                nvox=np.sum(mask)
                tstat=np.mean(vol_fd[mask])
                vol_output = {'file' : fn_base, 'roi' : roi, 'corrtype' : corr, 'hemi' : hemi, 'cb_idx' : val, 'nvoxels' : nvox, 't' : f'{tstat:.2f}' }
                output.append(vol_output)
save_dict_to_csv(output,output_fn)
```
